```python
numeric_cols=list(data.select_dtypes(include=['int64','float64']))
categorical_cols=list(data.select_dtypes(include=['object']))
print(numeric_cols)
print(len(categorical_cols))
```

```python
cat_1=['Marital Status','T Stage ','N Stage','6th Stage','differentiate']
cat_2=[]
for i in categorical_cols:
  if i not in cat_1:
    cat_2.append(i)
```

```python
print(cat_1)
print(cat_2)
```

```python
data[cat_1]
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(data[cat_1])

data_transformed_1 = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out())

data_transformed_1
```

```python
data[cat_2]
```

```python
data_transformed_2 = pd.DataFrame()
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in cat_2:
  data_transformed_2[col] = le.fit_transform(data[col])

data_transformed_2
```

```python
data[numeric_cols]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(data[numeric_cols])

data_transformed_3 = pd.DataFrame(scaler.transform(data[numeric_cols]), columns=numeric_cols)

data_transformed_3
```

```python
data_transformed_1
```

```python
data_transformed_2
```

```python
data_transformed_3
```

```python
data_transformed = pd.concat([data_transformed_1, data_transformed_2, data_transformed_3], axis=1)
data_transformed
```
